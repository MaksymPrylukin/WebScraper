```python
import os
import pandas as pd
from icalendar import Calendar, Event
```

Calendar export

```python
group_name = input("Enter yoour group name: ")
```

```python
classes = pd.read_csv(f"schedules/{group_name}.csv", encoding="UTF-8")
```

```python
calendar = Calendar()
calendar["dtstart"] = "TZID=Europe/Warsaw:20260223T000000"
calendar["dtend"] = "TZID=Europe/Warsaw:20260930T235999"
calendar["summary"] = f"Schedule for {group_name} group"
calendar["PRODID"] = "-//UEK//Apollo to iCalendar export//EN"
```

```python
for index, row in classes.iterrows():
    event = Event()
    event["dtstart"] = "TZID=Europe/Warsaw:"+row["Termin"].replace("-", "")+"T"+row["Start time"].replace(":", "")+"00"
    event["dtend"] = "TZID=Europe/Warsaw:"+row["Termin"].replace("-", "")+"T"+row["End time"].replace(":", "")+"00"
```
